### Data Transfomation

```python
cruise_ship_df['Tonnage'] = raw_data_df['Tonnage'] * 1000
cruise_ship_df['length'] = raw_data_df['length'] * 100
cruise_ship_df['cabins'] = raw_data_df['cabins'] * 100
cruise_ship_df['crew'] = raw_data_df['crew'] * 100
cruise_ship_df['passengers'] = raw_data_df['passengers'] * 100
```

### 4. What Cruise Line has the highest crew to passenger ratio?

```python
new_cruise_ship_df = cruise_ship_df.copy()
```

```python
type(new_cruise_ship_df.groupby('Cruise_line')[['crew']].sum())
```

```python
cruise_line_total_crew_df = new_cruise_ship_df.groupby('Cruise_line')[['crew']].sum()
```

```python
cruise_line_total_crew_df
```

```python
cruise_line_total_passenger_df = new_cruise_ship_df.groupby('Cruise_line')[['passengers']].sum()
```

```python
cruise_line_total_passenger_df
```

```python
cruise_line_crew_passenger_df = pd.concat([cruise_line_total_crew_df, cruise_line_total_passenger_df], axis=1)
```

```python
cruise_line_crew_passenger_df
```

```python
cruise_line_crew_passenger_df['crew_passenger_ratio'] = cruise_line_crew_passenger_df['passengers'] / cruise_line_crew_passenger_df['crew']
```

```python
cruise_line_crew_passenger_df.sort_values(by='crew_passenger_ratio')
```

```python
cruise_line_crew_passenger_df[cruise_line_crew_passenger_df['crew_passenger_ratio']==cruise_line_crew_passenger_df['crew_passenger_ratio'].max()]
```
